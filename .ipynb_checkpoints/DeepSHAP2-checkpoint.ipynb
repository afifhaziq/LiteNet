```python
import numpy as np
import torch
from model import NtCNN

sequence = 37
features = 20
learning_rate = 0.001
batch_size = 64
num_class = 10
num_epoch = 30

device = "cuda" if torch.cuda.is_available() else "cpu"
model = NtCNN(sequence=sequence, features=features, num_class=num_class).to(device)
model.load_state_dict(torch.load("saved_dict//NtCNN_MALAYAGT_740Features_best_model.pth"))
model.eval()

train = np.load("dataset/MALAYAGT/train.npy", allow_pickle=True)

train = torch.from_numpy(train.astype(np.float32))

train[:, [12,13,14,15,16,17,18,19]] = 0

x_train = train[:, :-1]
y_train = train[:, -1]
```

```python
import shap

shap.initjs()

random_indices = np.random.choice(len(x_train), size=100, replace=False)

# Select the first n samples as background
background = x_train[random_indices].to(device)

# Initialize DeepExplainer
explainer = shap.DeepExplainer(model, background)
print('start calculating shap')

# Compute SHAP values with additivity check disabled
shap_value = explainer.shap_values(background, check_additivity=False)
print('done')

print(type(shap_value))
```

```python
type(shap_value)
```

```python
# Let's examine the shape of SHAP values and understand what we got
print("SHAP values type:", type(shap_value))
print("SHAP values shape:", shap_value.shape)
print("Expected shape: (background_samples, features, num_classes)")

# Check the background data shape
print("Background data shape:", background.shape)

# Extract dimensions
num_samples, num_features, num_classes = shap_value.shape
print(f"Number of samples: {num_samples}")
print(f"Number of features: {num_features}")
print(f"Number of classes: {num_classes}")
```

```python
# Create a more comprehensive visualization
# SHAP values shape: (background_samples, features, num_classes)

# Calculate mean absolute SHAP values across all classes
# This gives us the overall importance of each feature
mean_abs_shap = np.abs(shap_value).mean(axis=2)  # Shape: (num_samples, features)

print("Mean absolute SHAP values shape:", mean_abs_shap.shape)

# For the summary plot, we need feature names
feature_names = [f'Byte {i}' for i in range(740)]


# Summary plot showing feature importance
shap.summary_plot(mean_abs_shap, background.cpu().numpy(), feature_names=feature_names, max_display=20)
```

```python
shap.summary_plot(mean_abs_shap, background.cpu().numpy(), feature_names=feature_names, max_display=20, plot_type="bar")
```

```python
# Save the results for future use
# This replicates what you were doing in your original notebook

# Calculate global SHAP values (similar to your original approach)
# Shape: (samples, features, classes) -> average over samples and classes
shap_value_global = np.abs(shap_value).mean(axis=(0, 2))  # Shape: (features,)

# Get the most important indices (top 20 features)
most_important_indices = np.argsort(shap_value_global)[::-1][:20]


print("Top 20 most important feature indices:", most_important_indices)
print("Shape of most_important_indices:", most_important_indices.shape)


# Save the global SHAP values
np.save('top740featuresMALAYAGT_DeepSHAP.npy', shap_value_global)
print("Saved global SHAP values to 'top740featuresMALAYAGT_DeepSHAP.npy'")

# Also save the most important indices
np.save('top20features_indices_MALAYAGT_DeepSHAP.npy', most_important_indices)
print("Saved top 20 feature indices to 'top20features_indices_MALAYAGT_DeepSHAP.npy'")
```
